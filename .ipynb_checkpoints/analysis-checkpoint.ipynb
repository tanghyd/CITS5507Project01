```python
quick_data = (
    df.loc[
        (df.Algorithm == 'quick') &
#         (df.Cutoff == 0) &
        (df.Construct == 'tasks'),
        ['Threads', 'Cutoff', 'Size', 'Duration']
    ]
    .groupby(['Threads', 'Cutoff', 'Size'])['Duration']
    .agg({'mean', 'min', 'max', 'std'})
)

quick_data
```

```python

# fig = plt.figure(figsize=(20, 12))
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16,6), sharex=True, sharey=True)

for i, cutoff in enumerate(quick_data.index.get_level_values('Cutoff').unique()):
    ax = axes[i]
    ax.set_title(f'Quick Sort Runtime vs. Array Size (Cutoff={cutoff})')

    
    # plot data
    for j, n_threads in enumerate(quick_data.index.get_level_values('Threads').unique()):
        data = quick_data.loc[
            (quick_data.index.get_level_values('Threads') == n_threads) &
            (quick_data.index.get_level_values('Cutoff') == cutoff)
        ]
        
        array_sizes = data.index.get_level_values('Size').values

        color=list(mcolors.TABLEAU_COLORS)[j]
        ax.plot(array_sizes, data['mean'].values, label=f'{n_threads} Threads', color=color)
        ax.errorbar(array_sizes, data['mean'].values, yerr=data['std'].values , fmt='o', markersize=4, color=color)
    
    if i == 0:
        ax.set_ylabel('Runtime (s)')
    ax.legend(loc='upper left')
        
    ax.set_xlabel('Array Size')
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.grid()

fig.suptitle('Sorting Algorithm Runtimes in C (Log-Log Scale)')
fig.tight_layout()
fig.show()
```
